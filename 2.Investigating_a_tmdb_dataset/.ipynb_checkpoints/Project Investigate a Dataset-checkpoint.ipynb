```python
# Exploratory Data Analysis
```

```python
# DIY labels to operate on data
labels = ['id', 'imdb_id', 'popularity', 'budget', 'revenue', 'original_title', 'cast', 'homepage', 'director', 'tagline', 'keywords', 'overview', 'runtime', 'genres', 'production_companies', 'release_date', 'vote_count', 'vote_average','release_year', 'budget_adj','revenue_adj']
df = pd.read_csv('tmdb-movies.csv', names = labels)
```

```python
df.head()
```

```python
df['homepage'].fillna('Other', inplace = True)
df['tagline'].fillna('Other', inplace = True)
df['keywords'].fillna('Other', inplace = True)
df['production_companies'].fillna('Other', inplace = True)
df['genres'].fillna('Other', inplace = True)
df['cast'].fillna('Other', inplace = True)
df['overview'].fillna('Other', inplace = True)
df['director'].fillna('Other', inplace =True)

df.info()
```

```python
df['budget'] = df['budget'].convert_objects(convert_numeric=True)
df['revenue'] = df['revenue'].convert_objects(convert_numeric=True)
df['popularity'] = df['popularity'].convert_objects(convert_numeric=True)
df['runtime'] = df['runtime'].convert_objects(convert_numeric=True)
df['vote_count'] = df['vote_count'].convert_objects(convert_numeric=True)
df['vote_average'] = df['vote_average'].convert_objects(convert_numeric=True)
df['release_eyar'] = df['release_year'].convert_objects(convert_numeric=True)
df['budget_adj'] = df['budget_adj'].convert_objects(convert_numeric=True)
df['revenue_adj'] = df['revenue_adj'].convert_objects(convert_numeric=True)

df.info()
```
