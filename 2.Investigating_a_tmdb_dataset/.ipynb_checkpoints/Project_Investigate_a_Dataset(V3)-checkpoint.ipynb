### 2.2 Data Cleaning

#### 2.2.1 On problem 1: missing values
   1. imdb_id
   2. cast
   3. homepage

#### 2.2.1.1 On filling in values or dropping
  - I was going to fill in those missing values with some non-null values. 
  - My goal is to check popularities and movies, so for those labels with which not cause a big difference on analysis, I will fill in those missing values with "Others".
  
 

```python
df_NaN_id = df[df['imdb_id'].isnull()]
df_NaN_id.head(n=5)
```

```python
df_NaN_cast = df[df['cast'].isnull()]
df_NaN_cast.head(n = 5)
```

```python
df['homepage'].fillna('Others', inplace = True)
df['tagline'].fillna('Others', inplace = True)
df['keywords'].fillna('Others', inplace = True)
df['production_companies'].fillna('Others', inplace = True)
df['overview'].fillna('Others', inplace = True)
df['cast'].fillna('Others', inplace = True)
df['genres'].fillna('Others', inplace = True)

df.info()
```

```python
df.tail(n = 4)
```

#### 2.2.2 On problem 2: wrong datetime format

```python
df['release_date_new'] = df['release_date'].str[:-2] + df['release_year'].apply(str)

df.tail(n=4)
```

```python
df.info()
```

```python
df['release_date_new'] = pd.to_datetime(df['release_date_new'])
df.tail()
```

### 3.2 Research Question 2  What kinds of properties are associated with movies that have high revenues?
 - I need to create a new data frame that contains correlated labels and drop unrelated labels.
 - Correlated labels:
     - original_title. Movies' names.
     - genres. Because certain genres might be popular for a certain period of time.
     - director. Because genius directors create classic movies and revenues.
     - Cast. Because people would like watch a movie because of a ceratin actor or actress.
     - popularity. Because this is what most of people chose and paid.
     - budget_adj. Because a higher investment may lead to a better quality of a movie.
     - vote_average. Because the vote average shows a relative rating of how people rate a movie. 
     - release_year. Because the more recent, the more common for people watching movies. 
     - runtime. Because the longer runtime might need larger budget which may contribute higher revenues.
     - vote_count. Because the more votes a movies receives, the more revenue a movie might be watched.
     - revenue_adj. This is the target value. 
 - Unrelated labels:
     - id. Because this is unrelated to the revenue.
     - budget. Because budget_adj is more accurate after adjusted inflation.
     - revenue. Because revenue_adj is more accurate after adjusted inflation.
     - runtime. Because this is unrelated to the revenue.
